## Importing of Libraries

```python
from selenium import webdriver
from selenium.webdriver.support.ui import Select
import pandas as pd
from selenium.webdriver.common.by import By
#from selenium.webdriver import ActionChains
import time
driver = webdriver.Chrome(executable_path='C:\drivers\chromedriver.exe')
driver.get('https://www.justia.com')
driver.maximize_window()
```

## Scraping of Single page

```python
lawyer_name = []
small_bio = []
phone_number = []
address = []
specialization = []
university = []
email_link = []


for result in results:
    lawyer_name.append(result.find_element(By.XPATH, value='.//a[@data-vars-action="OrganicListingProfileName"]').text)
    small_bio.append(result.find_element(By.XPATH, value='.//div[contains(@class, "lawyer-expl")]').text)
    try:
        phone_number.append(result.find_element(By.XPATH, value='.//a[@data-vars-action="OrganicListingPhone"]').text)
    except:
        phone_number.append('NAN')
    address.append(result.find_element(By.XPATH, value= './/span[contains(@class, "-address")]').text.replace('\n', ''))
    try:
        specialization.append(result.find_element(By.XPATH, value= './/span[contains(@class, "-practices")]').text)
    except:
        specialization.append('NAN')
    try:
        university.append(result.find_element(By.XPATH, value='.//span[contains(@class, "-law-schools ")]').text)
    except:
        university.append('NAN')
    try:
        email_link.append(result.find_element(By.XPATH, value='.//a[@data-vars-action="OrganicListingEmail"]').get_attribute('href'))
    except:
        email_link.append('NAN')
```

```python
lawyers_df = pd.DataFrame({'LawyerName':lawyer_name, 'SmallBio':small_bio, 'PhoneNumber':phone_number,
                                    'Address':address, 'University':university, 'EmailLink':email_link})
```

## Scraping of Multiple pages

```python
lawyer_name = []
small_bio = []
phone_number = []
address = []
specialization = []
university = []
email_link = []

for i in range(1, 21):
    driver.get("https://www.justia.com/lawyers/texas/texas-city?page=" + str(i))
    results = driver.find_elements(By.XPATH, value='//div[@data-vars-action="OrganicListing"]')


    for result in results:
        # Name
        
        lawyer_name.append(result.find_element(By.XPATH, value='.//a[@data-vars-action="OrganicListingProfileName"]').text)
        
        # Small_bio
        
        small_bio.append(result.find_element(By.XPATH, value='.//div[contains(@class, "lawyer-expl")]').text)
        
        # Phone number
        try:
            phone_number.append(result.find_element(By.XPATH, value='.//a[@data-vars-action="OrganicListingPhone"]').text)
        except:
            phone_number.append('NAN')
        
        # Address
        address.append(result.find_element(By.XPATH, value= './/span[contains(@class, "-address")]').text.replace('\n', ''))
        
        # Specialisation
        try:
            specialization.append(result.find_element(By.XPATH, value= './/span[contains(@class, "-practices")]').text)
        except:
            specialization.append('NAN')
        
        # University
        try:
            university.append(result.find_element(By.XPATH, value='.//span[contains(@class, "-law-schools ")]').text)
        except:
            university.append('NAN')
        
        # Email link
        try:
            email_link.append(result.find_element(By.XPATH, value='.//a[@data-vars-action="OrganicListingEmail"]').get_attribute('href'))
        except:
            email_link.append('NAN')

            
            
lawyers_multiple_df = pd.DataFrame({'LawyerName':lawyer_name, 'SmallBio':small_bio, 'PhoneNumber':phone_number,
                                    'Address':address, 'University':university, 'Specialisation': specialization, 'EmailLink':email_link})
lawyers_multiple_df
```
